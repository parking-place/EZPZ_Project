```python
import json
import pandas as pd
import os
```

```python
# JSON 파일이 저장된 폴더 경로
files_path = '/Users/gyeonmunju/Desktop/PlayData/도서자료 요약/Training/[원천]도서요약_train/사회과학'

# 저장할 CSV 파일 이름
output_file = f'output.csv'

# JSON 파일 데이터를 읽어들이기
data = []
for filename in os.listdir(files_path):
    if filename.endswith('.json'):
        with open(os.path.join(files_path, filename), 'r', encoding='utf-8') as file:
            json_data = json.load(file)
            data.append(json_data)

# JSON 데이터를 CSV 파일로 변환
dataframe = pd.json_normalize(data)
dataframe.to_csv(output_file, index=False, encoding='utf-8', escapechar='\\')

# 결과 확인
dataframe = pd.read_csv(output_file)
print(dataframe)
```
